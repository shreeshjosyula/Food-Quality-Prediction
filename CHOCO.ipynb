```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

```

```python
choco = pd.read_csv('C:\\Machine Learning\\flavors_of_cacao.csv')
```

```python
choco.columns=['company','bean-origin','ref','reviewdate','cocoapercent','companylocation','rating','beantype','broadbeanorigin']
```

```python
choco.info()
```

```python
label_quality = LabelEncoder()
choco['company'] = label_quality.fit_transform(choco['company'])
choco['bean-origin'] = label_quality.fit_transform(choco['bean-origin'])
choco['companylocation'] = label_quality.fit_transform(choco['companylocation'])
```

```python
choco=choco.drop('beantype',axis=1)
choco=choco.drop('broadbeanorigin',axis=1)
```

```python
def slicing(percent):
    numeral=str(percent)
    numeral = numeral[:2]
    return numeral
    
```

```python
choco['cocoapercent'] = choco['cocoapercent'].apply(slicing)
```

```python
choco['cocoapercent']=choco['cocoapercent'].astype(int)
```

```python
choco.head(100)
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categorical_features = [0])
choco = onehotencoder.fit_transform(choco)
```
